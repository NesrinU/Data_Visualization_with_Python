## Liniendiagramm (Serien/Datenrahmen)

- Was ist ein Liniendiagramm und warum wird es verwendet?

    - Ein Liniendiagramm oder Liniendiagramm ist ein Diagramm, das Informationen als eine Reihe von Datenpunkten anzeigt, die als „Marker“ bezeichnet werden und durch gerade Liniensegmente verbunden sind. 
    - Es ist ein grundlegender Diagrammtyp, der in vielen Bereichen üblich ist.
    - Verwenden Sie das Liniendiagramm, wenn Sie einen kontinuierlichen Datensatz haben. Diese eignen sich am besten für trendbasierte Visualisierungen von Daten über einen bestimmten Zeitraum.

## Zusatzliche Übungen

-Matplotlib - ``.subplot2grid()``-Funktion

- Diese Funktion bietet mehr Flexibilität beim Erstellen eines Achsenobjekts an einer bestimmten Position des Rasters(grid). 
- Es ermöglicht auch, dass sich das Achsenobjekt über mehrere Zeilen oder Spalten aufgespannt ist.

``plt.subplot2grid (shape, location, rowspan, colspan)``

- Im folgenden Beispiel wird ein 3X3-Raster des Abbildungsobjekts mit Achsenobjekten unterschiedlicher Größe in Zeilen- und Spaltenspannen gefüllt, die jeweils eine andere Darstellung zeigen.

    - https://www.tutorialspoint.com/matplotlib/matplotlib_subplot2grid_function.htm
    - https://medium.com/kodluyoruz/matplotlibde-%C3%A7oklu-%C3%A7izim-nas%C4%B1l-yap%C4%B1l%C4%B1r-subplot-gridspec-b33e2077727c
    - https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplot2grid.html


```python
x = np.arange(1, 10)

plt.figure(figsize=(15, 5))
plt.suptitle("LineStyles & Markers & Markersize & Color", fontsize=16, y = 1.1)

plt.subplot(2, 2, 1)
plt.plot(x, np.exp(x))

plt.subplot(2, 2, 2)
plt.plot(x, x*x)

plt.subplot(2, 2, 3)
plt.plot(x, np.log(x));
```

```python
# rowspan and colspan
plt.figure()
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)
ax2 = plt.subplot2grid((3, 3), (0, 2), rowspan=3)
ax3 = plt.subplot2grid((3, 3), (1, 0), rowspan=2, colspan=2)

x = np.arange(1, 10)

ax1.plot(x, np.exp(x))
ax1.set_title('exp, colspan=2')

ax2.plot(x, x*x)
ax2.set_title('square, rowspan=3')

ax3.plot(x, np.log(x))
ax3.set_title('log, rowspan=2, colspan=2')

plt.tight_layout();
```

## Einige andere Diagramme
- Neben dem standardmäßigen Liniendiagramm sind viele andere Zeichenstile verfügbar, auf die alle zugegriffen werden kann, indem das Schlüsselwort kind an ``plot()`` übergeben wird. Die vollständige Liste der verfügbaren Grundstücke lautet wie folgt:

    -   `bar` für vertikale Balkendiagramme
    -   `barh` für horizontale Balkendiagramme
    -   `hist` for Histogram
    -   `box` for Boxplot
    -   `kde` oder `density` für Dichtediagramme
    -   `area` für Flächengrundstücke
    -   `pie` für Tortendiagramme
    -   `scatter` für Streudiagramme
    -   `hexbin` für hexbin plot

## Manueller Abstand auf Subplots ()
Verwenden Sie ``.subplots_adjust``, um den Abstand manuell anzupassen.

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 6), sharey=True, sharex=True)

ax[0, 0].plot(alter, gehalt, "r"), ax[0, 0].set_title("Gehalt nach Alter")
ax[0, 1].plot(alter, gehalt_2, "m"), ax[0, 1].set_title("Gehalt_2 nach Alter")
ax[1, 0].plot(alter, gehalt, "g"), ax[1, 0].set_title("Gehalt nach Alter")
ax[1, 1].plot(alter, gehalt_2, "y"), ax[1, 1].set_title("Gehalt_2 nach Alter")

fig.subplots_adjust(left=None,
                    bottom=None,
                    right=None,
                    top=None,
                    wspace=0.3,
                    hspace=0.5)
# plt.tight_layout()  # versuchen Sie auch tight_layout()
```
